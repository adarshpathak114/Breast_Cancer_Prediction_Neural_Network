#Breast cancer prediction on NN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.shape
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.describe()
```

1 --> Benign

0 --> Malignant

Separating the features and target


```python
X = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

## Standardize the model

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_standard = scaler.fit_transform(X_train)
X_test_standard = scaler.transform(X_test)
```

```python
# importing tensorflow keras
import tensorflow as tf
# to get same accuracy result at every run we used random seed it is same as random_state
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# building neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train_standard, y_train, epochs = 10, validation_split=0.1)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

##Accuracy of model on test data

```python
loss, accuracy =model.evaluate(X_test_standard, y_test)
print(accuracy)
```

```python
y_pred = model.predict(X_test_standard)
```

```python
print(y_pred)
# this is probability so we will convert to 0 or 1 for good classification using argmax
```

model.predict() gives the prediction probability of each class for that data point:

```python
# argmax function
my_list = [0.25, 0.67]
index_of_max_value = np.argmax(my_list)
print(my_list)
print(index_of_max_value)
# we will apply same logic in every data points in y_pred to get 0 $ 1 classification
```

```python
# converting the prediction probabilities in to class labels
y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```
